```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame(data)

# Ensure 'Dataset Size' and 'Number of Cores' are integers for easy manipulation
df['Dataset Size'] = df['Dataset Size'].apply(lambda x: int(x.replace('k', '000').replace('K', '000')))
df['Number of Cores'] = df['Number of Cores'].astype(int)

# The baseline time for speedup calculation is the execution time with a single process
baseline_df = df[df['Number of Processes'] == 1].copy()
baseline_df = baseline_df.set_index(['Number of Cores', 'Dataset Size'])

# Calculate Speedup as the ratio of baseline execution time to the current execution time
# Here, the baseline time is defined for each combination of 'Number of Cores' and 'Dataset Size' when only one process is used
df['Speedup'] = df.apply(lambda row: baseline_df.loc[(row['Number of Cores'], row['Dataset Size']), 'Time (Seconds)'] / row['Time (Seconds)'], axis=1)

# Calculate Efficiency as Speedup divided by the Number of Processes (not the Number of Cores)
df['Efficiency'] = df['Speedup'] / df['Number of Processes']


# Define custom palette
palette = sns.color_palette("tab10")  # This is a predefined palette with distinct colors.

# Plotting
# Speedup plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Number of Processes', y='Speedup', hue='Number of Cores', style='Dataset Size', markers=True, palette=palette)
plt.title('Speedup by Number of Processes and Cores')
plt.xlabel('Number of Processes')
plt.ylabel('Speedup')
plt.legend(title='Cores and Dataset Size')
plt.grid(True)
plt.show()

# Efficiency plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Number of Processes', y='Efficiency', hue='Number of Cores', style='Dataset Size', markers=True, palette=palette)
plt.title('Efficiency by Number of Processes and Cores')
plt.xlabel('Number of Processes')
plt.ylabel('Efficiency')
plt.legend(title='Cores and Dataset Size')
plt.grid(True)
plt.show()
```
