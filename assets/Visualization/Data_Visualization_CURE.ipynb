```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/cluster_results.csv'
data = pd.read_csv(file_path)

# Visualize the clusters
plt.figure(figsize=(10, 6))
colors = {0: 'blue', 1: 'red', 2: 'yellow', -1: 'green'}
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=data.iloc[:, 2].map(colors), alpha=0.5)
plt.title('Cluster Visualization with Custom Colors')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
